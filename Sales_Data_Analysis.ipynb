```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_excel("/content/SampleSuperstore.xls")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dtypes
```

```python
numeric_df = df.select_dtypes(include=['number'])
numeric_df.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['Sales'] = df['Sales'].ffill()
plt.figure(figsize=(8,5))

sns.barplot(x='Product Category',y='Sales',data=df)

plt.title("Sales by Category")

plt.show()
```

```python
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
df['Profit'] = df['Profit'].ffill()
plt.figure(figsize=(8,5))

sns.barplot(x='Region',y='Profit',data=df)

plt.title("Profit by Region")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Sales'],bins=30)

plt.title("Sales Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Profit'])

plt.title("Profit Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='Sales',y='Profit',data=df)

plt.title("Sales vs Profit")

plt.show()
```

```python
top=df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

top.plot(kind="bar",figsize=(10,5))

plt.title("Top 10 Customers")

plt.show()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
df['Order Date']=df['Order Date'].ffill()

df['Month']=df['Order Date'].dt.month_name()

monthly=df.groupby('Month')['Sales'].sum()

monthly.plot(kind='line',marker='o',figsize=(10,5))

plt.title("Monthly Sales")

plt.show()
```

```python
df.groupby("Product Category")["Sales"].sum().plot(kind="pie",autopct="%1.1f%%",figsize=(6,6))

plt.ylabel("")

plt.title("Sales Share")

plt.show()
```

```python
sns.pairplot(df.select_dtypes(include='number'))
```

**Insights**

• Technology category generated highest sales.

• West region generated highest profit.

• Few customers contribute major sales.

• Profit and Sales have positive relationship.

• Some products show high sales but low profit.

• Dataset has minimal missing values.
